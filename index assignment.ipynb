```python
import pandas as pd
train= pd.read_csv("train.csv",index_col=0)
train
```

```python
# dropping columns with null values.
train.dropna(axis=1,inplace=True)
train
```

```python
# Feature selection
train.corr()["SalePrice"].sort_values(ascending=False).head(5)
```

```python
# Based on the correlation results we select the top two features.
X_train=train[["OverallQual","GrLivArea"]]
X_train
```

```python
# Target Feature
y_train=train["SalePrice"]
y_train
```

```python
# One hot encoding for the Train features
from sklearn.preprocessing import OneHotEncoder

```

```python
ohe = OneHotEncoder(drop='first', sparse=False)
category_feature = ["OverallQual"]
X_train_category = X_train.loc[:, category_feature]
X_train_ohe = pd.DataFrame(ohe.fit_transform(X_train_category),
                           index=X_train.index)
X_train_ohe.head()
```

```python
# X_train with OverallQual dropped and concat with ohe
X_train=X_train.drop(columns=category_feature)
X_train_tr=pd.concat([X_train,X_train_ohe],axis=1)
X_train_tr.head()
```

```python
# Import test dataset.

test= pd.read_csv("test.csv",index_col=0)
test
```

```python
# dropping columns with null values.
test.dropna(axis=1,inplace=True)
test
```

```python
# Based on the correlation results under train we select the top two features.
X_test=test[["OverallQual","GrLivArea"]]
X_test
```

```python
# One hot encoder for test features
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse=False)
category_feature = ["OverallQual"]
X_test_category = X_test.loc[:, category_feature]
X_test_ohe = pd.DataFrame(ohe.fit_transform(X_test[category_feature]),
                           index=X_test.index)
X_test_ohe.head()
```

```python
# X_test with OverallQual dropped and concat with ohe test
X_test=X_test.drop(columns=category_feature)
X_test_tr=pd.concat([X_test,X_test_ohe],axis=1)
X_test_tr.head()
```

```python
# convert feature names to strings so there is not a TypeError with sklearn

X_train_tr.columns = X_train_tr.columns.astype(str)
X_test_tr.columns = X_test_tr.columns.astype(str)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train_tr, y_train)

```

```python
y_hat_train = linreg.predict(X_train_tr)
y_hat_test = linreg.predict(X_test_tr)
```

```python
from sklearn.metrics import mean_squared_error
train_mse = mean_squared_error(y_train, y_hat_train)
train_mse
```
